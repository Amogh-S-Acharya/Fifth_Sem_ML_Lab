```python
import pandas as pd

df = pd.read_csv('/content/retail_sales_dataset.csv')
display(df.head())
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df['Month'] = df['Date'].dt.month
df = pd.get_dummies(df, columns=['Gender', 'Product Category'], drop_first=True)
X = df.drop(['Date', 'Transaction ID', 'Customer ID', 'Total Amount', 'Gender_Male', 'Product Category_Clothing', 'Product Category_Electronics'], axis=1)
y = df['Total Amount']
display(X.head())
display(y.head())
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
```

```python
from sklearn.model_selection import cross_val_score

scores = cross_val_score(model, X, y, cv=5)
```

```python
display(scores)
```

```python
import numpy as np

mean_score = np.mean(scores)
std_score = np.std(scores)

print(f"Mean cross-validation score: {mean_score:.4f}")
print(f"Standard deviation of cross-validation scores: {std_score:.4f}")
```

```python
# Example 1: A younger customer buying a larger quantity of a cheaper item
sample_input_1 = pd.DataFrame({
    'Age': [25],
    'Quantity': [5],
    'Price per Unit': [20],
    'Month': [9]
})

predicted_amount_1 = model.predict(sample_input_1)
print(f"Predicted Total Amount (Example 1): ${predicted_amount_1[0]:.2f}")

```

```python
# Create a sample input record (replace with your desired values)
# The columns must match the features used for training: 'Age', 'Quantity', 'Price per Unit', 'Month'
sample_input = pd.DataFrame({
    'Age': [35],
    'Quantity': [2],
    'Price per Unit': [100],
    'Month': [6]
})

# Fit the model on the entire dataset before making a prediction
model.fit(X, y)

# Make the prediction
predicted_amount = model.predict(sample_input)

# Display the predicted total amount
print(f"Predicted Total Amount: ${predicted_amount[0]:.2f}")
```
